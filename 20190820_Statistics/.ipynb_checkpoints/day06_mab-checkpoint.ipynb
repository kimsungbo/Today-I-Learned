### 에타 제곱

```python
전체평균 = sleep['extra'].mean()
```

```python
X = sleep['extra']
```

```python
import numpy
```

```python
전체SS = numpy.sum((X - 전체평균) ** 2)
```

```python
sleep.groupby('group').agg({'extra': 'mean'})
# sleep을 group 변수에 따라(by) 그룹을 지어(group)
# extra 컬럼을 평균(mean)내라
# agg: aggregate의 약자 (모아서 통계내다) 
```

```python
처치SS = (  # 그룹SS, 처치 = treatment
    10 * (0.75 - 전체평균) ** 2 + # 1번그룹의 10명의 사람
    10 * (2.33 - 전체평균) ** 2)
```

```python
에타제곱 = 처치SS / 전체SS
```

```python
에타제곱  # 베게 차이가 수면시간의 분산의 16.1%를 설명한다
```

### 코헨의 d

```python
코헨의d = (2.33 - 0.75) / sleep['extra'].std()
```

```python
코헨의d  # 두 집단이 0.78 표준편차 차이가 난다 (IQ 11점 정도 차이)
```

### 입실론 그리디 + 이동평균

```python
count = {'A': 0, 'B': 0}
value = {'A': 0, 'B': 0}
real = {'A': 10, 'B': 9}

for _ in range(1000):
    if uniform(0, 1) < epsilon:
        s = random.choice('AB')
    else:
        if value['A'] > value['B']:
            s = 'A'
        else:
            s = 'B'

    reward = normal(real[s], 1)
    count[s] += 1
    value[s] += (reward - value[s]) / count[s]
    print(time, s)
    
```

### 입실론 그리디 + 지수 이동 평균

```python
value = {'A': 0, 'B': 0}
real = {'A': 10, 'B': 9}
count = {'A': 0, 'B': 0}

for time in range(1000):
    if uniform(0, 1) < epsilon:
        s = random.choice('AB')
    else:
        if value['A'] > value['B']:
            s = 'A'
        else:
            s = 'B'

    reward = normal(real[s], 1)
    count[s] += 1
    value[s] += (reward - value[s]) / 10
    print(time, s)
    
    if time > 500:  # 시점이 500을 넘어갈 때 A의 가치가 7로 떨어진다
        real['A'] = 7
```
