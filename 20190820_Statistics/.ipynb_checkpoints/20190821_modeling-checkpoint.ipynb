## 리트리버 분포

```python
개판 = []
for i in range(100):
    한계 = 10      # parameter 1
    짜증확률 = .4  # parameter 2

    옐로카드 = 0
    기간 = 0
    while 옐로카드 < 한계:
        짜증 = binomial(1, 짜증확률)
        if 짜증 == 1:
            옐로카드 += 1 # 옐로카드 = 옐로카드 + 1
        else:
            옐로카드 = 0
        기간 += 1
    개판.append(기간)
seaborn.distplot(개판)
```

# 확률

## 독립

* 두 사건 A, B가 독립이면 P(A|B) = P(A) 이고 P(B|A) =P(B)
* P(A, B) = P(A|B) P(B) = P(A) P(B)

## 조건부 독립

B가 주어졌을때
* (A) → (B) → (C) : A 는 몰라도 C 알 수 있다
* P(A, C | B) = P(C | B) * P(B | A) * P(A) / P(B) = P(C | B) * P(A | B)


```python
연기력 = normal(50, 15, 100) # 평균 50, 표준편차 15인 배우 100명
```

```python
seaborn.distplot(연기력)
```

### 연기의 영향이 매출의 평균에 영향을 미치는 모형

```python
연기의영향 = 100 * 연기력 + 20000
매출 = normal(연기의영향, 800)
```

```python
seaborn.jointplot(연기력, 매출)
```

### 연기의 영향이 매출의 표준편차에 영향을 미치는 모형

```python
연기의영향 = 100 * 연기력
매출 = normal(20000, 연기의영향)
```

```python
seaborn.jointplot(연기력, 매출)
```

## 선형모형

* x에따라 y의 평균을 예측

```python
연기의영향 = 0.01 * 연기력 - 0.5

seaborn.distplot(연기의영향)
```

## 로지스틱 선형 모형

* 0 ~ 1 사이의 출력을 갖는 함수
* 확률은 0 ~ 1 사이의 값이므로, 확률을 출력하는 함수를 만들 수 있음

```python
from scipy.special import expit
```

```python
expit(0)
```

```python
expit(10)
```

```python
expit(-6)
```

```python
seaborn.distplot(expit(연기의영향))
```

```python
주연상 = binomial(5, expit(연기의영향))
```

```python
seaborn.distplot(주연상)
```

## 선형모형: 연속형 -> 이산형

```python
실력 = normal(50, 15, 100)
```

```python
실력의영향 = expit(0.5 * 실력 - 25)
```

```python
합격 = binomial(1, 실력의영향)
```

```python
seaborn.distplot(실력)
```

```python
seaborn.distplot(합격)
```

```python
# 결과 either 0 or 1
seaborn.jointplot(실력, 합격)
```

## 선형모형: 이산형 -> 연속형

```python
# 80%의 확률로 작동, 100대의 공기청정기
공기청정기 = binomial(1, .8, 100)
```

```python
청정기의영향 = 8 * 공기청정기 + 8  # 청정기가 안 돌때 8, 돌때 16
```

```python
공기질 = normal(청정기의영향, 30)
```

```python
seaborn.distplot(공기질)
```

## 로지스틱 선형모형: 이산형 -> 이산형

```python
공기청정기 = binomial(1, .8, 100)
청정기의영향 =  expit(2 * 공기청정기 - 1)
앨러지 = binomial(1, 청정기의영향)
```

아래처럼 if문을 쓰지 않아도 된다.

```python
if 공기청정기 == 1:
    청정기의영향 = 0.73
else:
    청정기의영향 = 0.26
```

```python
seaborn.distplot(앨러지)
```
