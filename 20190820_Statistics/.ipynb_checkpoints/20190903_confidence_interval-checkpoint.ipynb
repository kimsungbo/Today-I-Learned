# 신뢰구간

### Sampling Distribution

* 동일한 분포에서 같은 크기의 표본(sample)을 추출
* 각 표본에서 통계치를 추정
* 포본 분포 : 추정된 통계치의 분포
* sampling 분포 != sample 분포

```python
from numpy.random import uniform
```

```python
import numpy
```

```python
x = uniform(0, 100, 30) # 0에서 100사이에서 30개의 수를 고르게 뽑는다
x
```

```python
numpy.mean(x)  # x의 평균
```

```python
ms = []                     # 빈 리스트를 만든다
for _ in range(100000):
    x = uniform(0, 100, 30) # 데이터 x를 만든다(샘플링)
    m = numpy.median(x)       # x의 평균
    ms.append(m)            # 리스트에 평균 추가
```

```python
ms
```

```python
import seaborn
```

```python
seaborn.distplot(x)   # 하나의 샘플에서 데이터는 어떻게 퍼져 있는가?
```

```python
seaborn.distplot(ms)  # 샘플링할 때마다 평균이 어떻게 달라지는가?
```

모분포를 sampling 하면

* 모분포를 닮지만 sample 마다 조금씩 달라진다
* 평균의 Sampling Distribution 을 구하면 모분포의 평균

* 중간값의 Sampling Distribution 도 구할 수 있다

## 95% 신뢰구간

= 95% 확실하다

```python
numpy.quantile(ms, .025)  # 하위 2.5%
```

```python
numpy.quantile(ms, .975) # 하위 97.5% (상위 2.5%)
```

33.13 ~ 66.86 안에 평균이 있을 가능성이 95%

## 99% 신뢰구간

```python
numpy.quantile(ms, .005)  # 하위 0.5%
```

```python
numpy.quantile(ms, .995) # 하위 99.5% (상위 0.5%)
```

28.19 ~ 71.55안에 평균이 있을 가능성이 99%

### Bootstraping

* Sample 하나가 있을 때
    * Sample 이 모분포와 비슷하다고 가정한다
    * Sample 이 충분히 크면 모분포와 비슷할 확률이 크다고 본다
* 그 Sample 안에서 무작위로 데이터를 뽑아서 Sample 여러개를 만든다
* 거기에서 평균이 존재할 가능성이 95 % 인 구간을 찾는다

* 복잡한 분포도 다룰 수 있다

```python
from sklearn.utils import resample
```

```python
# Resample = 가지고 있는 Sample 에서 다시 Sample 을 뽑아준다
resample([3, 4, 5])
```

```python
ms = []
for _ in range(100000):
    x = resample(cars['speed'])  # 우리가 가진 샘플에서 다시 샘플링
    m = numpy.mean(x)            # 평균
    ms.append(m)                 # 리스트에 추가
```

```python
seaborn.distplot(ms)
```

### Bootstrapping으로 오차범위 구하기

```python
numpy.mean(ms)
```

```python
numpy.quantile(ms, .025)
```

```python
numpy.quantile(ms, .975)
```

```python
15.398784200000001 - 13.94
```

```python
15.398784200000001 - 16.84
```

신뢰구간

* 15.39 +- 1.4 (오차범위)
* 13.92 ~ 16.84

## 실습: 부트스트래핑으로 선형 모형의 파라미터의 95% 신뢰구간 구하기

```python
coefs = []
intercepts = []

for _ in range(10000):
    df = resample(cars)                   # 샘플링
    x = df[['speed']]                     # 독립변수
    y = df['dist']                        # 종속변수
    
    model = LinearRegression()            # 모형 만들기
    model.fit(x, y)                       # 파라미터 추정
    
    coefs.append(model.coef_)             # 추정한 계수를 리스트에 추가
    intercepts.append(model.intercept_)   #    "   절편을     "
```

```python
seaborn.distplot(coefs) 
```

```python
numpy.quantile(coefs, [.025, .975])
```

```python
numpy.quantile(intercepts, [.025, .975])
```

## 실습: sleep 데이터

sleep 데이터를 열어서 group으로 extra를 예측하는 선형 모형을 만들고
각 파라미터의 신뢰구간을 구해보세요.

```python
sleep = pandas.read_csv('sleep.csv')
```

```python
model = LinearRegression()
```

```python
x = sleep[['group']]
y = sleep['extra']
```

```python
model.fit(x, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
coefs = []
intercepts = []

for _ in range(10000):
    df = resample(sleep)
    
    x = df[['group']]
    y = df['extra']
    
    model = LinearRegression()
    model.fit(x, y)
    
    coefs.append(model.coef_)
    intercepts.append(model.intercept_)
```

```python
seaborn.distplot(coefs)
```

```python
seaborn.distplot(intercepts)
```

```python
numpy.quantile(coefs, [.025, .975])
```
