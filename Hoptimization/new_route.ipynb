```python
import folium
```

```python
import pandas as pd
```

```python
new_route = pd.read_csv('./new_route_latlon.csv')
```

```python
new_route.drop
```

```python
new_route.drop([0], inplace=True)
```

```python
Busline = folium.Map(location = [37.194263,126.8727078], zoom_start=11)

folium.LayerControl(collapsed=False).add_to(Busline)

for index, row in new_route.iterrows():
    folium.Marker([row['위도'], row['경도']], popup=row['정류소명'], icon = folium.Icon(icon='cloud')).add_to(Busline)
    
Busline
```
