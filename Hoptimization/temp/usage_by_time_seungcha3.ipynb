```python
def count(route_id, station_id, time, usage_type):
    
    time = str(int(float(time)))
    
    if(time[0:1] == '0'):
        time = time[1:2]
        
#     print(route_id, station_id, time, usage_type)
    
#     print(time in df_usage['시간대'])


    index = df_usage.index[(df_usage['노선ID'] == route_id)
                     & (df_usage['표준정류장ID'] == station_id)
                     & (df_usage['시간대'] == time)]
#     print(index)
#     print(df_usage.loc[index][usage_type])
    df_usage.set_value(index, usage_type, df_usage.loc[index][usage_type]+1 )
```
