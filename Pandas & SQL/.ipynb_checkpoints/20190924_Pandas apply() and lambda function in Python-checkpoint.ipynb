# 시도 / 시군구 나누기

```python
# train 테이블에 region 컬럼에 들어있는 값 중 10001번째 데이터를 호출
sample = train['region'][10000]

# 텍스트 데이터를 공백을 기준으로 잘라서 리스트에 저장
sample_split = sample.split()

# 저장된 리스트에서 첫번째 원소를 가져옴
sample_split[0]
```

```python
# split()
train['region'][10000].split()[0]
```

```python
def get_sido(x):
    return x.split()[0]
```

```python
def get_sigungu(x):
    return x.split()[1]
```

```python
get_sido(sample)
```

```python
get_sigungu(sample)
```

```python
list(train['region'].items())
```

```python
sido = []
for i, v in train['region'].items():
    sido.append(get_sido(v))
```

```python
sido
```

```python
train['sido'] = sido
```

```python
sigungu = []
for i, v in train['region'].items():
    sigungu.append(get_sigungu(v))
```

```python
sigungu
```

```python
train['sigungu'] = sigungu
```

```python
train
```

```python
train[train['sido'] == '제주']
```

# 한줄로

## apply()

```python
train['sido'] = train['region'].apply(get_sido)
```

## lambda

```python
get_sigungu = lambda x: x.split()[1]
```

```python
train['sigungu'] = train['region'].apply(get_sigungu)
```

# 더 짧게

```python
train['sigungu'] = train['region'].apply(lambda x: x.split()[1])
```

# Q1. 컬럼 transacted_date를 참조하여 결제가 있었던 연도를 뽑아내 year라는 칼럼에 저장하세요.

```python
def get_year(x):
    return x.split('-')[0]
```

```python
year = []
for i, v in train['transacted_date'].items():
    year.append(get_year(v))
```

```python
train['year'] = year
```

```python
train
```

```python
train['year'] = train['transacted_date'].apply(lambda x: x.split('-')[0])
```

```python
train
```

# Q2. 컬럼 transacted_date를 참조하여 결제가 있었던 월을 뽑아내 month라는 칼럼에 저장하세요.

```python
train['month'] = train['transacted_date'].apply(lambda x: x.split('-')[1])
```
