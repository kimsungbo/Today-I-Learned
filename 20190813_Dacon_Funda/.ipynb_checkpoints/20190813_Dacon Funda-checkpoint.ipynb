### 92일동안 월, 화, 수, 목, 금, 토, 일 갯수를 구해서 평균

```python
missing_data['transacted_date'] = pd.to_datetime(missing_data['transacted_date'])
```

```python
missing_data['day_of_week'] = missing_data['transacted_date'].dt.weekday_name
```

```python
missing_data
```

```python
grouped = missing_data.groupby(['store_id', 'day_of_week'])['amount'].mean().to_frame()
```

```python
grouped.head(100)
```

```python
dates = pd.DataFrame (pd.date_range(start = '20190301', end='20190531'))
```

```python
dates.columns = ['day_of_week']
```

```python
dates
```

```python
dates['day_of_week'] = dates['day_of_week'].dt.weekday_name
```

```python
num_days = dates.groupby(['day_of_week']).size().to_frame()
```

```python
num_days.columns=['num_days']
```

```python
num_days
```

```python
grouped = grouped.reset_index()
```

```python
grouped
```

```python
merge_result = pd.merge(grouped, num_days['num_days'],on='day_of_week',how='left')
```

```python
merge_result
```

```python
merge_result['sales'] = merge_result['amount'] * merge_result['num_days']
```

```python
merge_result
```

```python
result = merge_result.groupby('store_id')['sales'].sum()
```

```python
result
```

```python
baseline.groupby(['store_id'])['amount'].sum().append(result).to_csv('./added_missing_stores_jan_to_feb.csv')
```
