## 가성비 선수 찾기 (공격형 미드필더 LW, LAM, CAM, RAM, RW)

### Potential 은 높으나 현재 Overall이 낮은 선수

Potential과 Overall의 차이를 사용하여 가능성 등급은 높은데 전체 등급이 낮은 선수를 찾은다음 
해당 포지션의 overall 1점당 wage 평균 보다 적은 wage 를 받는 선수를 찾아보자.

```python
fifa['difference'] = (fifa['Potential'] - (fifa['Overall']))
```

```python
fifa['difference'].unique()
```

```python
def growth(d):
    if d == 0:
        return "Maintain"
    elif d >=1 and d<=9:
        return "Small"
    elif d >=10 and d<=18:
        return "Medium"
    elif d >19:
        return "Big"
```

```python
fifa['growth'] = fifa['difference'].apply(growth)
```

```python
#Overall 1점당 Wage
fifa['wage_per_overall'] = fifa['Wage'] / fifa['Overall']
```

```python
# 포지션별 overall 1점당 wage 평균
overall_wage_by_position = pd.DataFrame(fifa.groupby('Position')['wage_per_overall'].mean())
overall_wage_by_position['wage_per_overall'] = overall_wage_by_position['wage_per_overall'].astype(float)
overall_wage_by_position
```

### LW (Left Wing)

LW 의 포지션의 경우 아래 선수들이 가성비가 높은 선수에 해당함
* E.Smith Rowe
* G. Middleton
* S.Abifade

```python
# Position LW인 선수 중에서 growth 'big' 인 경우
fifa.loc[(fifa['growth']== 'Big') & (fifa['Position']== 'LW')].sort_values(by='Potential', ascending=False)
```

```python
overall_wage_by_position.loc['LW', 'wage_per_overall']
```

```python
# Position LW인 선수 중에서 growth 'big' 인 경우 & overall 1점당 wage 가 해당 포지션의 overall 1점당 wage의 평균 보다 작은경우
LW_growth = fifa.loc[(fifa['growth']== 'Big') & (fifa['Position']== 'LW') & (fifa['wage_per_overall'] < overall_wage_by_position.loc['LW', 'wage_per_overall'])].sort_values(by='Potential', ascending=False)
LW_growth
```

### 다른포지션의 선수이지만 가능성이 높은 선수들을 영입해오자!

RW / LW 의 포지션의 경우 아래의 weight로 potential 계산

* 16% Crossing
* 12% Attack Positioning
* 11% Dribbling
* 11% Ball Control
* 10% Shot Power
* 10% Long Shots
* 10% Reactions
* 06% Short Passing
* 05% Heading
* 05% Vision
* 04% Acceleration
* 04% Sprint Speed

그럼 위의 weight로 계산한 potential이 현재 포지션의 potential보다 높고, LW/RW의 평균 Wage보다 낮은 Wage를 받는 선수를 영입했을시 이득!
